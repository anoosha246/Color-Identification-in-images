```python

####################### Basic Information #######################

#Main Libraries used:
    #1. cv2 (for loading the image and converting into RGB format)
    #2. matplotlib (for plotting)
    #3. KMeans (for clustering)
    #4. squarify (for plotting treemap)

#Steps:
    #1. Read the image and convert into suitable format and then adjust its size.
    #2. Use KMeans to detect colors from the image.
    #3. Plotting a treemap for indicating the percentage of each color in the image and comparing it with the given image.
    
    
```

```python

####################### Including libraries #######################

import cv2
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.cluster import KMeans
import squarify
    
```

```python

####################### Including the function #######################

def clr_convrt(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

```python

####################### Reading the image #######################

img_bgr = cv2.imread("colors.png")

####################### converting img from BGR to RGB #######################

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python

####################### Modifying the image #######################



new_img = cv2.resize(img_rgb, (300, 300), interpolation = cv2.INTER_AREA)
new_img = new_img.reshape(new_img.shape[0]*new_img.shape[1], 3)

####################### using KMeans Clustering Algo #######################

clst = KMeans(n_clusters = 7)

counts = Counter(clst.fit_predict(new_img))
clr_centr = clst.cluster_centers_
order_centr = [clr_centr[i] for i in counts.keys()]
clrs = [clr_convrt(order_centr[i]) for i in counts.keys()]
```

```python

####################### plotting a Treemap #######################

%matplotlib inline
squarify.plot(counts.values(), label = clrs, color = clrs)

for i in range(1):
    plt.show()
    plt.imshow(img_rgb)  
    
```
